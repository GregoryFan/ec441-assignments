1. A network using Pure ALOHA has an offered load of
$G = 0.6$ 

a. Compute the throughput of the network.

b. Compute the throughput of the network if it were using Slotted ALOHA instead.

Answer:

a. Throughput of network:

$S = Ge^{-2G}$

$S = 0.6 * e ^ {-2 * 0.6}$

$S \approx 0.181$

The throughput of the Pure ALOHA network is 0.181.



b. Throughput in terms of Slotted ALOHA:

$S = Ge^{-G}$

$S = 0.6 * e ^ {-0.6}$

$S \approx 0.329$

The throughput of the Slotted ALOHA network is 0.329.

3. Why is CSMA/CD no longer used on modern ethernet?

Answer: CSMA/CD is no longer used on modern ethernet as it has switched to using switches and a full-duplex system. Meaning there is no possibility of collision.
