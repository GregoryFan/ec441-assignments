#### 1. A fiber link has an attenuation of 0.2 dB/km. The link length is 50 km.

##### (a) Compute the total attenuation in dB.
##### (b) If the transmitted power is 10 mW, what is the received power (in dBm)?

Answers:

a. 0.2 * 50 = 10dB

b. 10mW = 10log(10) = 10dBm\
10dBm - 10dBm = 0dBm

#### 5. For binary signaling in an AWGN channel, the probability of error is given by:
$
P_e = Q\left(\frac{A}{\sigma}\right)
$

#### Suppose that:
$
\frac{A}{\sigma} = 2
$

##### (a) Is the error probability high or low?
##### (b) What happens to the error probability if $A$ increases?

Answers:

a. Q(2) is relatively high, Q(3) on the slides is already around 1 in 1000, which is quite a lot.

b. Increasing A decreases the amount covered by the integral defined in Q, thus BER gets lower.
